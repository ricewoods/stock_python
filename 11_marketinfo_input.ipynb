```python
import numpy as np
import pandas as pd
import datetime as dt
from pandas_datareader import DataReader
from pandas import Series, DataFrame
```

```python
filedate = '20220419'
```

```python
# 予想のために作成したinputデータ（予想したい日の前日）
df_inputai = pd.read_csv('data/inputAI_' + filedate + '.csv', index_col=0, header=[0], parse_dates=[0])
```

```python
# マーケット情報をInput
df_market = pd.read_csv('data/market_' + filedate + '.csv', index_col=0, header=None, skiprows=2, thousands=',')
```

```python
df_market.replace({'--':0, '--%':0}, inplace=True)
```

```python
df_market.replace({'%':''}, regex=True, inplace=True)
```

```python
df_inputai['key']  = 1
```

```python
# AIへのinputデータ元作成
df_inputai2 = pd.merge(df_inputai, df_market, left_on='key', right_index=True)
```

```python
pd.set_option('display.max_rows', 60) 
# defult:60,制限なし:None
pd.set_option('display.max_columns', None) 
# defult:20,制限なし:None
df_inputai2
```

```python
# すべて数値形式に変換
df_inputai3 = df_inputai2.astype(float, errors = 'raise')
```

```python
df_inputai3.to_csv('data/inputAI2_' + filedate + '.csv')
```
