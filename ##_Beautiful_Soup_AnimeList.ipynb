```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://lnovel.jp/anime/now/'
```

```python
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.content, "html.parser")
```

```python
titles = []
details = []

for i in soup.find_all(['h3', 'details']):
    if i.name == 'h3' and i.get("class") == ['anime_title']:
        titles.append(i.get_text(strip=True))
        
    elif i.name == "details" and i.get("class") == ['mb-1']:
        details.append(list(i.stripped_strings))
```

```python
import pandas as pd
import numpy as np
df_titles = pd.DataFrame(titles)
df_titles.columns = ['Titles']
df_details = pd.DataFrame(details)
```

```python
df_list1 = pd.concat([df_titles, df_details], axis=1)
```

```python
pd.set_option('display.max_rows', None) 
# defult:60,制限なし:None
pd.set_option('display.max_columns', None) 
# defult:20,制限なし:None
df_list1
```

```python
# 先頭の列をコピー
index_column = df_list1.iloc[:, 0:2]
index_column.columns = ['Title', 'Summary']
```

```python
# 残りの列を取得
remaining_columns = df_list1.iloc[:, 2:]
```

```python
# 残りの列を行に変換
melted_df = pd.melt(df_list1, id_vars=['Titles'], var_name='Variable', value_name='Value')
```
