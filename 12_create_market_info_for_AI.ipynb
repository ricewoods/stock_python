```python
import numpy as np
import pandas as pd
import datetime as dt
from pandas_datareader import DataReader
from pandas import Series, DataFrame
```

```python
# 機械学習用データ作成（予想したい日の前日分までの累積）
filedate = '20220304'
```

```python
# AI用データを今日からデータがなくなるまで読み込みしてコンカチ
# これまでにストックしたデータ(前日分)
df_mk01 = pd.read_csv('data/market/AUD_JPY 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk02 = pd.read_csv('data/market/EUR_JPY 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk03 = pd.read_csv('data/market/EUR_USD 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk04 = pd.read_csv('data/market/GBP_JPY 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk05 = pd.read_csv('data/market/TRY_JPY 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk06 = pd.read_csv('data/market/USD_JPY 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk07 = pd.read_csv('data/market/NYダウ平均株価 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk08 = pd.read_csv('data/market/日経平均株価 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk09 = pd.read_csv('data/market/日経225先物 先物の過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk10 = pd.read_csv('data/market/オーストラリアS&P_ASX 200 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk11 = pd.read_csv('data/market/ドイツDAX指数 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
df_mk12 = pd.read_csv('data/market/米ドル指数先物 過去データ.csv', index_col=0, header=[0], parse_dates=[0])
```

```python
df_market = pd.concat([df_mk01, df_mk02, df_mk03, df_mk04, df_mk05, df_mk06, df_mk07,
                      df_mk08, df_mk09, df_mk10, df_mk11, df_mk12], axis=1)
```

```python
# データチェック
# df_market.isnull().any() {'年':'', '月':'', '日':''}
# PERをobject属性から'float64'属性に変更
df_market2 = df_market.replace({'%':'', ',':'', 'K':000, 'M':000000, 'B':000000000, '-$':0}, regex=True)
df_market2 = df_market2.astype('float64')
```

```python
df_market2.rename(index = {'2022年03月11日':'20220311', '2022年03月10日':'20220310',
                           '2022年03月09日':'20220309', '2022年03月08日':'20220308',
                           '2022年03月07日':'20220307', '2022年03月04日':'20220304',
                           '2022年03月03日':'20220303', '2022年03月02日':'20220302'}, inplace=True)
```

```python
# 予想のために作成したinputデータ（予想したい日の前日）
df_inputai = pd.read_csv('data/inputAI_' + filedate + '.csv', index_col=0, header=[0], parse_dates=[0])
```

```python
df_inputai['date'] = filedate
```

```python
# AIへのinputデータ元作成
df_inputai2 = pd.merge(df_inputai, df_market2, left_on='date', right_index=True)
```

```python
df_inputai2 = df_inputai2.replace(np.nan,0)
df_inputai2
```
