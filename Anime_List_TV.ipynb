```python
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
import tempfile
```

```python
chrome_options = Options()
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--headless=new")

# 毎回ユニークなユーザーデータディレクトリを作る
chrome_options.add_argument(f"--user-data-dir={tempfile.mkdtemp()}")
driver = webdriver.Chrome(options=chrome_options)

```

```python
# 対象ページへアクセス
driver.get("https://bangumi.org/genres/anime")

# 地域セレクトボックスを取得（PC版のID）
# select = Select(driver.find_element("id", "pcAreaSection"))
# PC用ではなくモバイル用を取得
select = Select(driver.find_element("id", "spAreaSection"))

# 例: 神奈川 (value="45") を選択
select.select_by_value("45")

# ページの内容が更新されるのを待つ
time.sleep(3)

# ページ内容をBeautifulSoupで解析
soup = BeautifulSoup(driver.page_source, "html.parser")

#--テストロジック Start--
# タイトル例を取得
#titles = [p.text for p in soup.select("p.program_supplement")]
#print(titles[:10])  # 最初の10件だけ表示

# ブラウザを閉じる
#driver.quit()
#--テストロジック End--
```

```python
# CSS セレクターを使って どの週の番組表になるのかを全て取得する
what_days = [n.get_text() for n in soup.select('div.h3_box h3')]
print(what_days)
```

```python
# CSS セレクターを使って アニメタイトルを全て取得する
# program_titles = [n.get_text() for n in soup.select('p.program_title, p.program_supplement')]
program_titles = [n.get_text() for n in soup.select('p.program_title')]
print(program_titles)
```

```python
# CSS セレクターを使って アニメの放送日時を全て取得する
program_supplement = [n.get_text() for n in soup.select('p.program_supplement')]
print(program_supplement)
```

```python
program_supplements = [s.replace('\u3000', ',') for s in program_supplement]
```

```python
import pandas as pd
import numpy as np
df_titles = pd.DataFrame(program_titles)
df_titles.columns = ['Title']
df_details = pd.DataFrame(program_supplements)
```

```python
df_list1 = pd.concat([df_titles, df_details], axis=1)
```

```python
pd.set_option('display.max_rows', 60)
# defult:60,制限なし:None
pd.set_option('display.max_columns', 20)
# defult:20,制限なし:None
df_list1
```

```python
df_list1[['日時', '放送局']] = df_list1[0].str.split(',', expand=True)
```

```python
df_list1
```

```python
#不要な列を削除
df_anime_list = df_list1.drop(df_list1.columns[1], axis=1)
```
