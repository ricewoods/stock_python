```python
import numpy as np
import pandas as pd
import datetime as dt
from pandas_datareader import DataReader
from pandas import Series, DataFrame
```

```python
# 機械学習用データ作成（予想したい日の前日分までの累積）
filedateP = '20220421'
filedateC = '20220422'
```

```python
# 前日分データ
df_p = pd.read_csv('data/inputAI2_' + filedateP + '.csv', index_col=0, header=[0], parse_dates=[0])
```

```python
# 当日分データ
df_c = pd.read_csv('data/inputAI2_' + filedateC + '.csv', index_col=0, header=[0], parse_dates=[0])
```

```python
# DataFrameをIndexでマージ
df_t2 = pd.merge(df_p, df_c, left_index=True, right_index=True, how='inner')
```

```python
# これまでにストックしたAIACデータ(前日まで)
df_t1 = pd.read_csv('data/inputAIAC2_' + filedateP + '.csv', index_col=0, header=[0], parse_dates=[0])
```

```python
# 上記データをコンカチ
# df_inputAI = pd.concat([df_t1, df_t2, df_t3, df_t4, df_t5, df_t6])
df_inputAI = pd.concat([df_t1, df_t2])
```

```python
# データチェック
df_inputAI.isnull().any()
```

```python
df_inputAI.to_csv('data/inputAIAC2_' + filedateC + '.csv')
```
