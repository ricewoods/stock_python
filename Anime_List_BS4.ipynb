```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bs4.jp/anime/'
```

```python
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.content, "html.parser")
```

```python
# CSS セレクターを使って アニメタイトルを全て取得する
# program_titles = [n.get_text() for n in soup.select('p.program_title, p.program_supplement')]
program_titles = [n.get_text() for n in soup.select("[class='common-list__tag tag--red'], [class='common-list__title'], [class='common-list__date']")]
print(program_titles)
```

```python
# CSS セレクターを使って アニメタイトルを全て取得する
# program_titles = [n.get_text() for n in soup.select('p.program_title, p.program_supplement')]
program_titles = [n.get_text() for n in soup.select("[class='common-list__title']")]
print(program_titles)
```

```python
# CSS セレクターを使って 放送日時を全て取得する
program_dates = [n.get_text() for n in soup.select("[class='common-list__date']")]
print(program_dates)
```

```python
import pandas as pd
import numpy as np
df_titles = pd.DataFrame(program_titles)
df_titles.columns = ['Title']
df_dates = pd.DataFrame(program_dates)
df_dates.columns = ['日時']
```

```python
df_list1 = pd.concat([df_titles, df_dates], axis=1)
```

```python
df_list1
```

```python
df_list1['放送局'] = "BS日テレ"
```

```python
df_list1
```

```python
df_list1 = df_list1.drop(index=df_list1.index[[8, 9]])
```

```python
url = 'https://www.bs4.jp/animenimuchoo/'
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
```

```python
# CSS セレクターを使って アニメタイトルを全て取得する
# program_titles = [n.get_text() for n in soup.select('p.program_title, p.program_supplement')]
program_titles2 = [n.get_text() for n in soup.select("[class='muchoo-day-box__list__item__text__name']")]
print(program_titles2)
```

```python
# CSS セレクターを使って 放送日時を全て取得する
program_dates2 = [n.get_text() for n in soup.select("[class='muchoo-day-box__list__item__text__data']")]
print(program_dates2)
```

```python
df_titles2 = pd.DataFrame(program_titles2)
df_titles2.columns = ['Title']
df_dates2 = pd.DataFrame(program_dates2)
df_dates2.columns = ['日時']
```

```python
df_list2 = pd.concat([df_titles2, df_dates2], axis=1)
```

```python
df_list2
```

```python
df_list2['放送局'] = "BS日テレ"
```

```python
df_list2
```

```python
df_bs4 = pd.concat([df_list1, df_list2])
```

```python
df_bs4 = df_bs4.reset_index(drop=True)
```
