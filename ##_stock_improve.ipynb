```python
# 予想前日の株価データ
df_stock = pd.read_csv('data/stock_20220423.csv', index_col=0, header=[0,1], parse_dates=[0])
```

```python
df_stock1 = df_stock.drop(df_stock.index[-1])
```

```python
# indexの順番を入れ替えて、ソート
df_stock1 = df_stock.swaplevel('Attributes','Symbols', axis=1)
df_stock1.sort_values(by=['Symbols', 'Attributes'], axis=1, inplace=True)
```

```python
# 直近5日分のデータを取得して行と列を入れ替え
df_stock2 = df_stock1.tail(5)
```

```python
# indexを再配番
df_stock2 = df_stock2.reset_index()
```

```python
# 日付行を削除し、行と列を入れ替え
df_stock2 = df_stock2.drop(columns=['Date']).T
```

```python
# インデックスを株価コードに、カラムをそれぞれの価格に変更
df_stock3 = df_stock2.pivot_table(index='Symbols', columns='Attributes')
```

```python
df_stock3
```

```python
df_stock3 = df_stock3.swaplevel(0,1, axis=1)
```

```python
# マルチカラムをリセット
df_stock3 = df_stock3.T.reset_index(drop=False).T
```

```python
# 既存のカラムを削除
df_stock4 = df_stock3.drop(df_stock3.index[[0,1]])
```
