```python
# 01通期財務データ（出典:https://irbank.net/download）
df_balance = pd.read_csv('data/finance/fy-balance-sheet.csv', index_col=0, header=[1], parse_dates=[0])
```

```python
# 02通期CFデータ（出典:https://irbank.net/download）
df_cashflow = pd.read_csv('data/finance/fy-cash-flow-statement.csv', index_col=0, header=[1], parse_dates=[0])
```

```python
# 03通期業績データ（出典:https://irbank.net/download）
df_profit_fy = pd.read_csv('data/finance/fy-profit-and-loss.csv', index_col=0, header=[1], parse_dates=[0])
```

```python
# 04通期配当データ（出典:https://irbank.net/download）
df_dividend = pd.read_csv('data/finance/fy-stock-dividend.csv', index_col=0, header=[1], parse_dates=[0])
```

```python
print(df_balance.shape)
print(df_cashflow.shape)
print(df_profit_fy.shape)
print(df_dividend.shape)
# print(df_netsales.shape)
# print(df_operating.shape)
# print(df_ordinary.shape)
# print(df_profit_pp.shape)
```

```python
# データの結合
df_stock = pd.concat([df_balance, df_cashflow, df_profit_fy, df_dividend
#          ,df_netsales, df_operating, df_ordinary, df_profit_pp
                     ],axis='columns')
```

```python
pd.set_option('display.max_rows', 60) 
# defult:60,制限なし:None
pd.set_option('display.max_columns', None) 
# defult:20,制限なし:None
df_stock
```

```python
# "年度"列を削除※複数ある場合、一斉に削除
df_stock.drop(columns=['年度'], inplace=True)
```

```python
# "-"をNaNに置換する
df_stock2 = df_stock.replace({'-':np.nan})
# すべて数値形式に変換
df_stock3 = df_stock2.astype(float, errors = 'raise')
```

```python
# データチェック(Nullがあるか否か)
# df_stock3.isnull().any()
df_stock3.drop(df_stock3.index[df_stock3['総資産'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['株主資本'].isnull()]
# NaNがある行を削除
df_stock3.drop(df_stock3.index[df_stock3['株主資本'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['利益剰余金'].isnull()]
# NaNがある行を削除
df_stock3.drop(df_stock3.index[df_stock3['利益剰余金'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['短期借入金'].isnull()]
# NaNを0に置換
df_stock3['短期借入金'] = df_stock3['短期借入金'].replace(np.nan,0)
```

```python
# df_stock3.index[df_stock3['長期借入金'].isnull()]
# NaNを0に置換
df_stock3['長期借入金'] = df_stock3['長期借入金'].replace(np.nan,0)
```

```python
# df_stock3.index[df_stock3['営業CF'].isnull()]
# NaNがある行を削除
df_stock3.drop(df_stock3.index[df_stock3['営業CF'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['投資CF'].isnull()]
# NaNがある行を削除
df_stock3.drop(df_stock3.index[df_stock3['投資CF'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['財務CF'].isnull()]
# NaNがある行を削除
df_stock3.drop(df_stock3.index[df_stock3['財務CF'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['設備投資'].isnull()]
# NaNを0に置換
df_stock3['設備投資'] = df_stock3['設備投資'].replace(np.nan,0)
```

```python
# df_stock3.index[df_stock3['現金同等物'].isnull()]
# df_stock3.index[df_stock3['営業CFマージン'].isnull()]
# df_stock3.index[df_stock3['売上高'].isnull()]
# df_stock3.index[df_stock3['営業利益'].isnull()]
df_stock3['営業利益'] = df_stock3['営業利益'].fillna(df_stock3['売上高'] * (df_stock3['営業利益'].mean() / df_stock3['売上高'].mean()))
```

```python
# df_stock3.index[df_stock3['経常利益'].isnull()]
df_stock3['経常利益'] = df_stock3['経常利益'].fillna(df_stock3['売上高'] * (df_stock3['経常利益'].mean() / df_stock3['売上高'].mean()))
```

```python
# EPS,ROE,ROA
# df_stock3.index[df_stock3['EPS'].isnull()]
# df_stock3.index[df_stock3['ROE'].isnull()]
# NaNがある行を削除
df_stock3.drop(df_stock3.index[df_stock3['ROE'].isnull()], inplace=True)
```

```python
# df_stock3.index[df_stock3['ROA'].isnull()]
# df_stock3.index[df_stock3['一株配当'].isnull()]
# NaNを0に置換
df_stock3['一株配当'] = df_stock3['一株配当'].replace(np.nan,0)
```

```python
# df_stock3.index[df_stock3['剰余金の配当'].isnull()]
# NaNを0に置換
df_stock3['剰余金の配当'] = df_stock3['剰余金の配当'].replace(np.nan,0)
df_stock3['自社株買い'] = df_stock3['自社株買い'].replace(np.nan,0)
df_stock3['配当性向'] = df_stock3['配当性向'].replace(np.nan,0)
df_stock3['総還元性向'] = df_stock3['総還元性向'].replace(np.nan,0)
df_stock3['純資産配当率'] = df_stock3['純資産配当率'].replace(np.nan,0)
```
